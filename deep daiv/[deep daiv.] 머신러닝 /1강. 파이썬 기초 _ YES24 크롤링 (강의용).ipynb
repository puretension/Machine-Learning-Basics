### BeautifulSoup
우선 html 소스 코드를 불러와 BeautifulSoup으로 원하는 정보를 파싱해보도록 하겠습니다.

```python
# 필요한 라이브러리(패키지) 불러오기
import requests # 서버와 통신하여 html 소스 코드를 반환하는 라이브러리
from bs4 import BeautifulSoup # html 소스 코드를 분석하여 원하는 정보를 추출할 수 있게 도와주는 라이브러리
import time # 시간 관련 정보를 제어할 수 있는 라이브러리
```

```python
# 필요한 라이브러리(패키지) 불러오기
import requests # 서버와 통신하여 html 소스 코드를 반환하는 라이브러리
from bs4 import BeautifulSoup # html 소스 코드를 분석하여 원하는 정보를 추출할 수 있게 도와주는 라이브러리
import time # 시간 관련 정보를 제어할 수 있는 라이브러리
# Requests
url = 'https://www.yes24.com/24/category/bestseller?CategoryNumber=001&sumgb=09'
r = requests.get(url)
# print(r.text)
html = r.text

#
```

```python
# BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 원하는 태그 찾기
# soup.find('태그', attrs = {'속성 키1': '속성 값1', '속성 키2': '속성 값2'})
soup.find('meta', attrs = {'name': 'title'}).get('content') # key: value
# '속성 키'가 class인 경우 {} 생략 가능합니다. 여러 개를 쓰는 경우에는 딕셔너리 형태로 적어주셔야 합니다.
```

```python
# 원하는 정보 추출하기
# <a href="/Product/Goods/117014613">세이노의 가르침</a>
# title = soup.find('a', {'href' :'/Product/Goods/117014613'}).text
title = soup.find('td','goodsTxtInfo').a.text
link = soup.find('td','goodsTxtInfo').a.get('href')

print(title,link)

len(soup.find_all('td','goodsTxtInfo'))
```

```python
# find_all을 활용하여 페이지에서 원하는 모든 정보 추출하기
url_tag_list = soup.find_all('td', 'goodsTxtInfo')
url_tag_list
```

```python
# 반복문을 활용하여 url 추출하기
for url_tag in url_tag_list:
    print(url_tag.a.get('href'))
```

```python
# 반복문을 활용하여 url을 리스트에 저장하기
url_list = []
for url_tag in url_tag_list:
    url = url_tag.a.get('href')
    url_list.append(url)

url_list
```

```python
# map과 lambda 함수를 활용하여 리스트에 저장하기
url_list = list(map(lambda x: x.a.get('href'), url_tag_list))
print(url_list)
```

```python
# test_url 지정
test_url = url_list[0]
```

```python
# urlopen으로 html 소스 코드 불러오기
r = requests.get('https://www.yes24.com' + test_url)
html = r.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 제목 수집 및 출력
title = soup.find('h2', 'gd_name').text
print(title)
```

```python
# 본문 내용 수집 및 출력
content = soup.find('div', 'infoWrap_txtInner').text
print(content.strip())
```

```python
# 저자 수집
author = soup.find('span','gd_auth').text
print(author)
```

```python
# 수집 결과
result = title, content, author
result
```

```python
# 사용자 정의 함수로 html 소스 코드 불러오기
def get_html(url):
    r = requests.get('https://www.yes24.com' + url)
    html = r.text

    return html

get_html('/Product/Goods/117014613')
```

```python
data_list = []
for url in url_list:
    html = get_html(url)
    time.sleep(1)

    soup = BeautifulSoup(html, 'html.parser')
    title = soup.find('h2', 'gd_name').text
    content = soup.find('div', 'infoWrap_txtInner').text.strip()
    author = soup.find('span', 'gd_auth').a.text

    data = title, content, author
    data_list.append(data)

    time.sleep(1) # 필수
    print(title, author, sep = ': ')
```
