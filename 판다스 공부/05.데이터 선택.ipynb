## 5-2. loc
### 이름을 이용해서 원하는 Row에서 원하는 Col 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.loc[['2번','3번']]
```

```python
df.loc['1번','국어'] # Index, Column
```

```python
df.loc[['1번','2번'],['이름','영어','수학']]
```

```python
df.loc['1번':'5번', '국어':'사회'] # 1번부터 5번까지, 국어부터 사회까지 
```

## 5-3 iloc
### 위치를 이용해서 원하는 Row에서 원하는 Col 선택
#### loc? -> 이름을 이용해서 원하는 Row에서 원하는 Col 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
# df.iloc[0]
df.iloc[0:3] # 슬라이싱
```

```python
# df.iloc[0,[2,3]] # 원하는 Row에 원하는 Col(쉼표로 연결한다!)
# df.iloc[0:2,[1,2,3]]
df.iloc[4,2] 
df.iloc[0:5, 2:8]
```

## 5-4 조건(T/F)

```python
df['키'] >= 185
```

```python
filt = (df['키'] >= 185)
# df[filt]
df[filt]
```

### df.loc[row_sel, col_sel]

```python
# df.loc[df['키'] >= 185, '수학'] # 키 185 이상 학생들의 수학 데이터만
df.loc[df['키'] >= 185, ['키','사회', '영어']]
```

### And와 OR 연산자

```python
df.loc[(df['키'] >= 185 )& (df['수학'] >= 80) & (df['SW특기'] == 'C')]
```

```python
df.loc[(df['국어'] >= 80) | (df['사회'] >= 80)]
```

### str 함수

```python
filt = df['이름'].str.startswith('송')
df[filt]
```

```python
filt = df['이름'].str.contains('태')
df[~filt]
```

```python
langs = ["Python","Java",'C']
filt = df['SW특기'].isin(langs)
df[filt]
```

```python
langs = ["python","java","c"]
filt = df['SW특기'].str.lower().isin(langs) # SW특기 Col의 데이터를 소문자로 바꾼뒤 langs에 있는지
df[filt]
```

```python
df['SW특기'].str.lower().isin(langs)
```

```python
df['SW특기'].str.contains('Java',na=True) # NaN 데이터에 대해서 True로 설정
```

```python
filt = df['SW특기'].str.contains('Java', na=False) # Java포함 & NAN 데이터 = False
df[filt]
```
