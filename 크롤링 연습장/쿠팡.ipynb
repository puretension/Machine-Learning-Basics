### 쿠팡 상품 크롤링

```python
import requests
from bs4 import BeautifulSoup

curPage = 1
url = 'https://www.coupang.com/np/search?rocketAll=false&searchId=7467206f6a1e439ba86adf6068a518ba&q=%EA%B2%8C%EC%9D%B4%EB%B0%8D%EB%A7%88%EC%9A%B0%EC%8A%A4&brand=&offerCondition=&filter=&availableDeliveryFilter=&filterType=&isPriceRange=false&priceRange=&minPrice=&maxPrice=&page=1&trcid=&traid=&filterSetByUser=true&channel=user&backgroundColor=&searchProductCount=119550&component=&rating=0&sorter=scoreDesc&listSize=36'
resp = requests.get(url,headers={'User-Agent' : 'Mozila/5.0'})
html = resp.text
soup = BeautifulSoup(html, 'html.parser')
products = soup.select('a.search-product-link')

for product in products:
    sub_url = 'https://www.coupang.com/' + product.attrs['href']
    resp = requests.get(sub_url, headers={'User-Agent' : 'Mozila/5.0'})
    html = resp.text
    soup = BeautifulSoup(html, 'html.parser')

    try:
        # 브랜드명
        brand_name = soup.select_one('a.prod-brand-name').text
    except:
        brand_name = ""

    # 상품명
    product_name = soup.select_one('h2.prod-buy-header__title').text

    # 가격 정보
    price = soup.select_one('span.total-price > strong').text # 자식 태그 접근
    
    print(brand_name, product_name, price)


```
